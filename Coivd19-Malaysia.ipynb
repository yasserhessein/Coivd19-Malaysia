```python

feature_names = ['cases_new', 'cases_import', 'cases_recovered', 'cases_active',
       'cases_cluster', 'cases_unvax', 'cases_pvax', 'cases_fvax',
       'cases_boost', 'cases_child', 'cases_adolescent', 'cases_adult',
       'cases_elderly', 'cases_0_4', 'cases_5_11', 'cases_12_17',
       'cases_18_29', 'cases_30_39', 'cases_40_49', 'cases_50_59',
       'cases_60_69', 'cases_70_79', 'cases_80', 'cluster_import',
       'cluster_religious', 'cluster_community', 'cluster_highRisk',
       'cluster_education', 'cluster_detentionCentre', 'cluster_workplace']
X = df[feature_names]
y = df['year']
y
```

```python
# Trend Analysis: Visualize the trend of new cases over time
import matplotlib.pyplot as plt
import seaborn as sns
# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Plot the trend of new cases over time
plt.figure(figsize=(14, 7))
sns.lineplot(x='date', y='cases_new', data=df)
plt.title('Trend of New COVID-19 Cases in Malaysia')
plt.xlabel('Date')
plt.ylabel('Number of New Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Vaccination Impact: Analyze the impact of vaccination on the number of cases

# Plot the number of cases for different vaccination statuses over time
plt.figure(figsize=(14, 7))
sns.lineplot(x='date', y='cases_unvax', data=df, label='Unvaccinated')
sns.lineplot(x='date', y='cases_pvax', data=df, label='Partially Vaccinated')
sns.lineplot(x='date', y='cases_fvax', data=df, label='Fully Vaccinated')
sns.lineplot(x='date', y='cases_boost', data=df, label='Boosted')
plt.title('COVID-19 Cases by Vaccination Status in Malaysia')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Age Group Analysis: Visualize the distribution of cases across different age groups

# Select relevant columns for age group analysis
age_group_columns = ['cases_0_4', 'cases_5_11', 'cases_12_17', 'cases_18_29', 'cases_30_39',
                     'cases_40_49', 'cases_50_59', 'cases_60_69', 'cases_70_79', 'cases_80']

# Sum the cases for each age group
age_group_cases = df[age_group_columns].sum()

# Plot the distribution of cases across different age groups
plt.figure(figsize=(14, 7))
age_group_cases.plot(kind='bar')
plt.title('Distribution of COVID-19 Cases Across Different Age Groups in Malaysia')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Cluster Analysis: Analyze the distribution of cases across different clusters

# Select relevant columns for cluster analysis
cluster_columns = ['cluster_import', 'cluster_religious', 'cluster_community', 'cluster_highRisk',
                   'cluster_education', 'cluster_detentionCentre', 'cluster_workplace']

# Sum the cases for each cluster
cluster_cases = df[cluster_columns].sum()

# Plot the distribution of cases across different clusters
plt.figure(figsize=(14, 7))
cluster_cases.plot(kind='bar')
plt.title('Distribution of COVID-19 Cases Across Different Clusters in Malaysia')
plt.xlabel('Cluster Type')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5,train_size=0.75)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# Import necessary libraries
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, HistGradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, precision_recall_fscore_support, accuracy_score

# Define classifiers
classifiers = {
    'SVM': SVC(kernel="linear", gamma='auto', random_state=10),
    'PassiveAggressive': PassiveAggressiveClassifier(random_state=10),
    'KNeighbors': KNeighborsClassifier(n_neighbors=36),
    'AdaBoost': AdaBoostClassifier(n_estimators=500,learning_rate=0.1),
    'DecisionTree': DecisionTreeClassifier(criterion="entropy", max_depth=3),
    'ExtraTrees': ExtraTreesClassifier(n_estimators=100, random_state=10),
    'RandomForest': RandomForestClassifier(n_estimators=1000, random_state=5, max_depth=2),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=10),
    'HistGradientBoosting': HistGradientBoostingClassifier(random_state=10),
    'MLP': MLPClassifier(random_state=10, max_iter=3000)
}

# Fit classifiers and print accuracy
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    print(f'Accuracy of {name} classifier on training set: {clf.score(X_train, y_train):.2f}')
    print(f'Accuracy of {name} classifier on test set: {clf.score(X_test, y_test):.2f}')
    print('')

# Predict y data with classifiers and evaluate
results = []
for name, clf in classifiers.items():
    y_pred = clf.predict(X_test)
    clf_rpt = [name] + [round(item * 100, 2) for item in precision_recall_fscore_support(y_test, y_pred, average='weighted') if item] + [round(accuracy_score(y_test, y_pred) * 100, 2)]
    results.append(clf_rpt)

# Print results
for result in results:
    print(result)
```
